```python
"dahyun+darwin=dahwin"
from requests_html import HTMLSession
import os
import requests
import random

s = HTMLSession()

for query in icon_list:
    endpage = 3
    curr_dir = os.getcwd()
    output = curr_dir + f"/{query}"
    
    if not os.path.exists(output):
        os.mkdir(output)
    
    urls = ["https://www.flaticon.com/search/{}?word={}&type=icon".format(x, query) for x in range(1, endpage)]
    for url in urls:
        print(url)
        r = s.get(url)
        content = r.html.find("div.icon--holder")
    
        for icon in content:
            icon_title = icon.find("img", first=True).attrs["alt"]
            icon_url = icon.find("img", first=True).attrs["src"]
            print(icon_title)
            print(icon_url)
    
            # Download and save the image
            image_content = requests.get(icon_url).content
    
            # Check if the file already exists
            image_filename = os.path.join(output, f"{icon_title}.png")
            random_number = ""
    
            while os.path.exists(image_filename):
                random_number = f"_{random.randint(1000, 9999)}"
                image_filename = os.path.join(output, f"{icon_title}{random_number}.png")
    
            with open(image_filename, "wb") as image_file:
                image_file.write(image_content)
    
            print(f"Saved: {image_filename}")
    

```
