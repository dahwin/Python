```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sine wave
t = np.linspace(0, 2 * np.pi, 1000)  # Time values from 0 to 2π
y = np.sin(t)

# Plot the sine wave
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a point (x, y)
x = 1
y = 0

# Rotate the point by 45 degrees counterclockwise
angle = np.pi / 4  # 45 degrees in radians
x_rotated = x * np.cos(angle) - y * np.sin(angle)
y_rotated = x * np.sin(angle) + y * np.cos(angle)

# Plot the original and rotated points
plt.plot([0, x], [0, y], label='Original Point')
plt.plot([0, x_rotated], [0, y_rotated], label='Rotated Point')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.axis('equal')  # Equal scaling for x and y axes
plt.title('Rotation in 2D')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate a simple harmonic oscillator
t = np.linspace(0, 10, 1000)  # Time values from 0 to 10 seconds
frequency = 2  # Frequency of oscillation (2 Hz)
amplitude = 1  # Amplitude of oscillation
y = amplitude * np.sin(2 * np.pi * frequency * t)

# Plot the oscillatory behavior
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Simple Harmonic Oscillator')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Pendulum parameters
length = 1.0  # Length of the pendulum arm
g = 9.81     # Acceleration due to gravity

# Time values
t_max = 10
num_frames = 200
t_vals = np.linspace(0, t_max, num_frames)

# Function to compute pendulum angle
def pendulum_angle(t):
    angular_frequency = np.sqrt(g / length)
    angle = np.pi / 4 * np.cos(angular_frequency * t)
    return angle

# Create figure and axis
fig, ax = plt.subplots()
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
pendulum_line, = ax.plot([], [], 'o-', lw=2)

# Animation update function
def update(frame):
    angle = pendulum_angle(t_vals[frame])
    x = length * np.sin(angle)
    y = -length * np.cos(angle)
    pendulum_line.set_data([0, x], [0, y])
    return pendulum_line,

# Create animation
ani = FuncAnimation(fig, update, frames=num_frames, blit=True, interval=50)

plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Pendulum Animation')
plt.grid()
plt.show()
```
